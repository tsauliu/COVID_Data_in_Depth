```python
#get daily covid cases

import warnings
warnings.simplefilter(action='ignore')

import pandas as pd
import os
df=pd.DataFrame()
for f in os.listdir(r'./baidu/'):
    d=pd.read_csv(r'./baidu/'+f)
    df=pd.concat([df,d])
df['抓取日期']=pd.to_datetime(df['抓取日期']).dt.date
df=df.drop_duplicates(subset=['城市','商场名称','抓取日期','抓取时段'])
df
```

```python
import datetime
c1=df['抓取日期']==datetime.date(2022,12,2)
city=df[c1]['城市'].drop_duplicates()
c2=df['城市'].isin(city)
df=df[c2]
```
